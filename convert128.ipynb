```python
import os
from PIL import Image
import glob

# Folder sumber dan tujuan
source_dir = "rafdb"
target_dir = "rafdb_128"
new_size = (128, 128)

# Ambil semua file gambar dari subfolder
image_paths = glob.glob(os.path.join(source_dir, "*", "*.*"))

# Format gambar yang diperbolehkan
valid_exts = (".png", ".jpg", ".jpeg", ".bmp")

for img_path in image_paths:
    if not img_path.lower().endswith(valid_exts):
        continue

    try:
        # Ambil nama subfolder dan nama file
        parts = img_path.split(os.sep)
        emotion_label = parts[-2]
        filename = parts[-1]

        # Buat path target dan pastikan folder tujuan ada
        target_folder = os.path.join(target_dir, emotion_label)
        os.makedirs(target_folder, exist_ok=True)

        # Resize dan simpan
        with Image.open(img_path) as img:
            img_resized = img.resize(new_size, Image.Resampling.LANCZOS)
            img_resized.save(os.path.join(target_folder, filename))

        print(f"✔️ {emotion_label}/{filename} disimpan di {target_folder}")

    except Exception as e:
        print(f"❌ Gagal memproses {img_path} - {e}")
```
